## Extract Diversity_CSV into DataFrames

```python
divers = "Resources/diversityindex.csv"
```

```python
divers_df=pd.read_csv(divers)
```

```python
divers_df.head()
```

## Transform Diversity DataFrame

```python
divers_df['County'],divers_df['State']=divers_df['Location'].str.rsplit(', ').str
```

```python
divers_df.head()
```

```python
divers_df_data = pd.DataFrame({
        'Diver_index':divers_df.groupby(['County','State'])['Diversity-Index'].mean(),
        'Black':divers_df.groupby(['County','State'])['Black or African American alone, percent, 2013'].mean(),
        'American':divers_df.groupby(['County','State'])['American Indian and Alaska Native alone, percent, 2013'].mean(),
        'Asian':divers_df.groupby(['County','State'])['Asian alone, percent, 2013'].mean(),
        'Hawaian':divers_df.groupby(['County', 'State'])['Native Hawaiian and Other Pacific Islander alone, percent,'].mean(),
        'Two or more races':divers_df.groupby(['County', 'State'])['Two or More Races, percent, 2013'].mean(),
        'Hispanic':divers_df.groupby(['County', 'State'])['Hispanic or Latino, percent, 2013'].mean(),
        'White':divers_df.groupby(['County', 'State'])['White alone, not Hispanic or Latino, percent, 2013'].mean()
    
})
```

```python
divers_df_data.head()
```
